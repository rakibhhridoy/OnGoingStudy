```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("data.csv")

fig = px.box(df, x= "Metals", y="Values", facet_col="Param", color="Age",
             boxmode="group", points='all')

fig.show()
```

```python
fig = px.box(df, x= "Metals", y="Values", facet_col="Param", color="Age",
             boxmode="group", points='all', facet_col_spacing= 0.002)
fig.show()
```

```python
sns.set(font_scale = 1.5)
sns.set_style("white")
sns.catplot(x='Metals', y='Values',
                 hue="Age",
                data=df, kind="box",
            height=6, aspect=1.3);
```

```python
sns.set(style="ticks", palette='Set2', font='Roboto Condensed')
sns.set_context("paper", font_scale=1.1, rc={"lines.linewidth": 1.1})
g=sns.factorplot(x="Metals", y="Values", hue="Age",
               col="Param", data=df, kind="box", height=4, aspect=0.5,
                 width=0.8,fliersize=2.5,linewidth=1.1, notch=False,orient="v")
sns.despine(trim=True)
```
