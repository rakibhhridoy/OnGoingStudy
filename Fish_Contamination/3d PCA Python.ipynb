```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

pca = pd.read_csv("pca_scores.csv")
pca.rename(columns = {'Unnamed: 0':'Elements'}, inplace = True)

arrayx = np.array([pca["Comp.1"]])
arrayy = np.array([pca["Comp.2"]])
arrayz = np.array([pca["Comp.3"]])

arrayx = arrayx.flatten()
arrayy = arrayy.flatten()
arrayz = arrayz.flatten()

labels = list(pca["Elements"])
x = preprocessing.normalize([arrayx])
y = preprocessing.normalize([arrayy])
z = preprocessing.normalize([arrayz])

x_reshape = x.reshape(14,)
y_reshape = y.reshape(14,)
z_reshape = z.reshape(14,)

ax3d = plt.figure().gca(projection='3d')

ax3d.scatter(x_reshape, y_reshape, z_reshape, c="black")

for x_label, y_label, z_label, label in zip(x_reshape, y_reshape, z_reshape, labels):
    ax3d.text(x_label, y_label, z_label, label)

ax3d.set_xlabel("PC1 (50.63%)")
ax3d.set_ylabel("PC2 (17.25%)")
ax3d.set_zlabel("PC3 (13.08%)")
plt.show()
plt.savefig("pca_3d.png", format="png", dpi=300)
```
