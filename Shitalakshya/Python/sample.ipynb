### Igeo

```python
background_values
```

```python
import math
```

```python
igeo_PbW = []
igeo_PbR = []

igeo_CrW = []
igeo_CrR = []

igeo_NiW = []
igeo_NiR = []

igeo_CuW = []
igeo_CuR = []

igeo_AsW = []
igeo_AsR = []

igeo_CdW = []
igeo_CdR = []

i = 0
while i < len(df["PbR"]):
    ig_Crr = math.log2(df["CrR"][i]/ (1.5 * background_values["Bm"][0]))
    igeo_CrR.append(ig_Crr)
    
    ig_Crw = math.log2(df["CrW"][i]/ (1.5 * background_values["Bm"][0]))
    igeo_CrW.append(ig_Crw)
    
    ig_Nir = math.log2(df["NiR"][i]/ (1.5 * background_values["Bm"][1]))
    igeo_NiR.append(ig_Nir)
    
    ig_Niw = math.log2(df["NiW"][i]/ (1.5 * background_values["Bm"][1]))
    igeo_NiW.append(ig_Niw)
    
    ig_Cur = math.log2(df["CuR"][i]/ (1.5 * background_values["Bm"][2]))
    igeo_CuR.append(ig_Cur)
    
    ig_Cuw = math.log2(df["CuW"][i]/ (1.5 * background_values["Bm"][2]))
    igeo_CuW.append(ig_Cuw)
    
    ig_Asr = math.log2(df["AsR"][i]/ (1.5 * background_values["Bm"][3]))
    igeo_AsR.append(ig_Asr)
    
    ig_Asw = math.log2(df["AsW"][i]/ (1.5 * background_values["Bm"][3]))
    igeo_AsW.append(ig_Asw)
    
    ig_Cdr = math.log2(df["CdR"][i]/ (1.5 * background_values["Bm"][4]))
    igeo_CdR.append(ig_Cdr)    
    
    ig_Cdw = math.log2(df["CdW"][i]/ (1.5 * background_values["Bm"][4]))
    igeo_CdW.append(ig_Cdw)    
    
    ig_pbr = math.log2(df["PbR"][i]/ (1.5 * background_values["Bm"][5]))
    igeo_PbR.append(ig_pbr)
    
    ig_pbw = math.log2(df["PbW"][i]/ (1.5 * background_values["Bm"][5]))
    igeo_PbW.append(ig_pbw)
    
    
    
    
    i += 1
```

```python
igeo_PbR, igeo_PbW
```

```python
igeo = pd.DataFrame({
    "CrW": igeo_CrW,
    "CrR": igeo_CrR,
    "NiW": igeo_NiW,
    "NiR": igeo_NiR,
    "CuW": igeo_CuW,
    "CuR": igeo_CuR,
    "AsW": igeo_AsW,
    "AsR": igeo_AsR,
    "CdW": igeo_CdW,
    "CdR": igeo_CdR,
    "PbW": igeo_PbW,
    "PbR": igeo_PbR
    
})
```

## CF CD PLI

```python
background_values
```

```python
cf_CrR = []
cf_CrW = []

cf_NiR = []
cf_NiW = []

cf_CuR = []
cf_CuW = []

cf_AsR = []
cf_AsW = []

cf_CdR = []
cf_CdW = []

cf_PbR = []
cf_PbW = []

i= 0
while i < len(df["CrW"]):
    val = df["CrW"][i]/ background_values["Bm"][0]
    cf_CrW.append(val)
    
    val = df["CrR"][i]/ background_values["Bm"][0]
    cf_CrR.append(val)
    
    val = df["NiR"][i]/ background_values["Bm"][1]
    cf_NiR.append(val)
    
    val = df["NiW"][i]/ background_values["Bm"][1]
    cf_NiW.append(val)
    
    val = df["CuW"][i]/ background_values["Bm"][2]
    cf_CuW.append(val)
    
    val = df["CuR"][i]/ background_values["Bm"][2]
    cf_CuR.append(val)
    
    val = df["AsR"][i]/ background_values["Bm"][3]
    cf_AsR.append(val)
    
    val = df["AsW"][i]/ background_values["Bm"][3]
    cf_AsW.append(val)
    
    val = df["CdW"][i]/ background_values["Bm"][4]
    cf_CdW.append(val)
    
    val = df["CdR"][i]/ background_values["Bm"][4]
    cf_CdR.append(val)
    
    val = df["PbR"][i]/ background_values["Bm"][5]
    cf_PbR.append(val)
    
    val = df["PbW"][i]/ background_values["Bm"][5]
    cf_PbW.append(val)
    
    
    
    i+=1 
```

```python
cf = pd.DataFrame({
    "CrW": cf_CrW,
    "CrR": cf_CrR,
    "NiW": cf_NiW,
    "NiR": cf_NiR,
    "CuW": cf_CuW,
    "CuR": cf_CuR,
    "AsW": cf_AsW,
    "AsR": cf_AsR,
    "CdW": cf_CdW,
    "CdR": cf_CdR,
    "PbW": cf_PbW,
    "PbR": cf_PbR
    
})

cf
```

```python
cf.describe()
```

```python
cf["Station"],cf["Location"],cf["Station_ID"] = stations["Stations"], stations["Location"], stations["Station_Code"]
cf
```

```python
cdR = []
cdW = []

i= 0
while i < len(df["CrW"]):
    val = cf_CrW[i] + cf_NiW[i] + cf_CuW[i] + cf_AsW[i] + cf_CdW[i] + cf_PbW[i]  
    cdW.append(val)
    
    val = cf_CrR[i] + cf_NiR[i] + cf_CuR[i] + cf_AsR[i] + cf_CdR[i] + cf_PbR[i]  
    cdR.append(val)

    i+=1 

cdR,cdW
```

```python
cf["CD_Winter"], cf["CD_Rainy"] = cdW, cdR
```

```python
cf
```

```python
PLI = []

i= 0
while i < len(df["CrW"]):
    val = (cf_CrW[i] * cf_NiW[i] * cf_CuW[i] * cf_AsW[i] * cf_CdW[i] * cf_PbW[i])**(1/6)  
    PLI.append(val)
    
    i+=1 

PLI
```

```python
cf["PLI"] = PLI
```

```python
cf
```

```python
cf.to_csv("CF.csv", sep=",")
cf.to_excel("CF.xlsx")
```

### EF

```python
background_values
```

```python
fe_conc = pd.read_csv("fe_conc.csv")
```

```python
fe_conc
```

```python
ef_CrR = []
ef_CrW = []

ef_NiR = []
ef_NiW = []

ef_CuR = []
ef_CuW = []

ef_AsR = []
ef_AsW = []

ef_CdR = []
ef_CdW = []

ef_PbR = []
ef_PbW = []

i= 0
while i < len(df["CrW"]):
    val = (df["CrW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][0]/background_values["Bm"][6])
    ef_CrW.append(val)
    
    val = (df["CrR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][0]/background_values["Bm"][6])
    ef_CrR.append(val)
    
    val = (df["NiR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][1]/background_values["Bm"][6])
    ef_NiR.append(val)
    
    val = (df["NiW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][1]/background_values["Bm"][6])
    ef_NiW.append(val)
    
    val = (df["CuW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][2]/background_values["Bm"][6])
    ef_CuW.append(val)
    
    val = (df["CuR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][2]/background_values["Bm"][6])
    ef_CuR.append(val)
    
    val = (df["AsR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][3]/background_values["Bm"][6])
    ef_AsR.append(val)
    
    val = (df["AsW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][3]/background_values["Bm"][6])
    ef_AsW.append(val)
    
    val = (df["CdW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][4]/background_values["Bm"][6])
    ef_CdW.append(val)
    
    val = (df["CdR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][4]/background_values["Bm"][6])
    ef_CdR.append(val)
    
    val = (df["PbR"][i] / fe_conc["Rainy"][i]) / (background_values["Bm"][5]/background_values["Bm"][6])
    ef_PbR.append(val)
    
    val = (df["PbW"][i] / fe_conc["Winter"][i]) / (background_values["Bm"][5]/background_values["Bm"][6])
    ef_PbW.append(val)
    
    
    
    i+=1 
```

## E and RI

```python
background_values
```

```python
acc_factor = df["PbR"]/background_values["Bm"][5]
acc_factor
```

```python
trv = pd.DataFrame({
    "Elements": ["Cr", "Ni", "Cu", "As", "Cd", "Pb"],
    "trv" : [26, 16, 16, 6, 0.6, 31]
})
```

```python
trv
```

```python
er_CrW = trv["trv"][0]* (df["CrW"]/background_values["Bm"][0])
er_CrR = trv["trv"][0]* (df["CrR"]/background_values["Bm"][0])

er_NiW = trv["trv"][1]* (df["NiW"]/background_values["Bm"][1])
er_NiR = trv["trv"][1]* (df["NiR"]/background_values["Bm"][1])

er_CuW = trv["trv"][2]* (df["CuW"]/background_values["Bm"][2])
er_CuR = trv["trv"][2]* (df["CuR"]/background_values["Bm"][2])

er_AsW = trv["trv"][3]* (df["AsW"]/background_values["Bm"][3])
er_AsR = trv["trv"][3]* (df["AsR"]/background_values["Bm"][3])

er_CdW = trv["trv"][4]* (df["CdW"]/background_values["Bm"][4])
er_CdR = trv["trv"][4]* (df["CdR"]/background_values["Bm"][4])

er_PbW = trv["trv"][5]* (df["PbW"]/background_values["Bm"][5])
er_PbR = trv["trv"][5]* (df["PbR"]/background_values["Bm"][5])
```

```python
er = pd.DataFrame({
    "CrW": er_CrW,
    "CrR": er_CrR,
    "NiW": er_NiW,
    "NiR": er_NiR,
    "CuW": er_CuW,
    "CuR": er_CuR,
    "AsW": er_AsW,
    "AsR": er_AsR,
    "CdW": er_CdW,
    "CdR": er_CdR,
    "PbW": er_PbW,
    "PbR": er_PbR
    
})

er
```

```python
RIW = er_CrW + er_NiW + er_CuW + er_AsW + er_CdW + er_PbW 
RIR = er_CrR + er_NiR + er_CuR + er_AsR + er_CdR + er_PbR 
```

```python
er["RI_Winter"], er["RI_Rainy"] = RIW, RIR
er
```

```python
er.describe()
```

```python
er.to_csv("RI.csv", sep=",")
er.to_excel("RI.xlsx")
```
