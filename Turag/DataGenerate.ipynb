```python
def data_gen(df):
    paramA = ["TempA", "pHA", "EC_A", "DO_A", "BOD_A", "TN_A"]
    paramB = ["TempB", "pHB", "EC_B", "DO_B", "BOD_B", "TN_B"]
    paramC = ["TempC", "pHC", "EC_C", "DO_C", "BOD_C", "TN_C"]
    paramD = ["TempD", "pHD", "EC_D", "DO_D", "BOD_D", "TN_D"]
    

    
    tempA_mat,tempB_mat,tempC_mat,tempD_mat,pHA_mat,pHB_mat,pHC_mat,pHD_mat,ECA_mat,ECB_mat,ECC_mat,ECD_mat,DOA_mat,DOB_mat,DOC_mat,DOD_mat,BODA_mat,BODB_mat,BODC_mat,BODD_mat,TNA_mat, TNB_mat, TNC_mat, TND_mat = val_append(df, paramA, paramB, paramC, paramD, i )
    
    return tempA_mat,tempB_mat,tempC_mat,tempD_mat,pHA_mat,pHB_mat,pHC_mat,pHD_mat,ECA_mat,ECB_mat,ECC_mat,ECD_mat,DOA_mat,DOB_mat,DOC_mat,DOD_mat,BODA_mat,BODB_mat,BODC_mat,BODD_mat,TNA_mat, TNB_mat, TNC_mat, TND_mat

        

        
        
def val_append(df, paramA, paramB, paramC, paramD, i):
    
    
    tempA_mat = []
    tempB_mat = []
    tempC_mat = []
    tempD_mat = []

    pHA_mat = []
    pHB_mat = []
    pHC_mat = []
    pHD_mat = []
    
    ECA_mat = []
    ECB_mat = []
    ECC_mat = []
    ECD_mat = []
    
    DOA_mat = []
    DOB_mat = []
    DOC_mat = []
    DOD_mat = []

    BODA_mat = []
    BODB_mat = []
    BODC_mat = []
    BODD_mat = []
    
    TNA_mat = []
    TNB_mat = []
    TNC_mat = []
    TND_mat = []

    
    i = 0
    while i < len(temp["TempA"]) - 1:
    
    #temp
    
        if df[paramA[0]][i] < df[paramA[0]][i+1]:
                val = np.random.uniform(df[paramA[0]][i+1], df[paramA[0]][i], 14)
                tempA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[0]][i], df[paramA[0]][i+1], 14)
            tempA_mat.append(val)

        if df[paramB[0]][i] < df[paramB[0]][i+1]:
                val = np.random.uniform(df[paramB[0]][i+1], df[paramB[0]][i], 14)
                tempB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[0]][i], df[paramB[0]][i+1], 14)
            tempB_mat.append(val)

        if df[paramC[0]][i] < df[paramC[0]][i+1]:
                val = np.random.uniform(df[paramC[0]][i+1], df[paramC[0]][i], 14)
                tempC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[0]][i], df[paramC[0]][i+1], 14)
            tempC_mat.append(val)

        if df[paramD[0]][i] < df[paramD[0]][i+1]:
                val = np.random.uniform(df[paramD[0]][i+1], df[paramD[0]][i], 14)
                tempD_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[0]][i], df[paramD[0]][i+1], 14)
            tempD_mat.append(val)


        #pH

        if df[paramA[1]][i] < df[paramA[1]][i+1]:
                val = np.random.uniform(df[paramA[1]][i+1], df[paramA[1]][i], 14)
                pHA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[1]][i], df[paramA[1]][i+1], 14)
            pHA_mat.append(val)

        if df[paramB[1]][i] < df[paramB[1]][i+1]:
                val = np.random.uniform(df[paramB[1]][i+1], df[paramB[1]][i], 14)
                pHB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[1]][i], df[paramB[1]][i+1], 14)
            pHB_mat.append(val)

        if df[paramC[1]][i] < df[paramC[1]][i+1]:
                val = np.random.uniform(df[paramC[1]][i+1], df[paramC[1]][i], 14)
                pHC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[1]][i], df[paramC[1]][i+1], 14)
            pHC_mat.append(val)

        if df[paramD[1]][i] < df[paramD[1]][i+1]:
                val = np.random.uniform(df[paramD[1]][i+1], df[paramD[1]][i], 14)
                pHD_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[1]][i], df[paramD[1]][i+1], 14)
            pHD_mat.append(val)

        #EC

        if df[paramA[2]][i] < df[paramA[2]][i+1]:
                val = np.random.uniform(df[paramA[2]][i+1], df[paramA[2]][i], 14)
                ECA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[2]][i], df[paramA[2]][i+1], 14)
            ECA_mat.append(val)

        if df[paramB[2]][i] < df[paramB[2]][i+1]:
                val = np.random.uniform(df[paramB[2]][i+1], df[paramB[2]][i], 14)
                ECB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[2]][i], df[paramB[2]][i+1], 14)
            ECB_mat.append(val)

        if df[paramC[2]][i] < df[paramC[2]][i+1]:
                val = np.random.uniform(df[paramC[2]][i+1], df[paramC[2]][i], 14)
                ECC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[2]][i], df[paramC[2]][i+1], 14)
            ECC_mat.append(val)

        if df[paramD[2]][i] < df[paramD[2]][i+1]:
                val = np.random.uniform(df[paramD[2]][i+1], df[paramD[2]][i], 14)
                ECD_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[2]][i], df[paramD[2]][i+1], 14)
            ECD_mat.append(val)


        #DO

        if df[paramA[3]][i] < df[paramA[3]][i+1]:
                val = np.random.uniform(df[paramA[3]][i+1], df[paramA[3]][i], 14)
                DOA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[3]][i], df[paramA[3]][i+1], 14)
            DOA_mat.append(val)

        if df[paramB[3]][i] < df[paramB[3]][i+1]:
                val = np.random.uniform(df[paramB[3]][i+1], df[paramB[3]][i], 14)
                DOB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[3]][i], df[paramB[3]][i+1], 14)
            DOB_mat.append(val)

        if df[paramC[3]][i] < df[paramC[3]][i+1]:
                val = np.random.uniform(df[paramC[3]][i+1], df[paramC[3]][i], 14)
                DOC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[3]][i], df[paramC[3]][i+1], 14)
            DOC_mat.append(val)

        if df[paramD[3]][i] < df[paramD[3]][i+1]:
                val = np.random.uniform(df[paramD[3]][i+1], df[paramD[3]][i], 14)
                DOD_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[3]][i], df[paramD[3]][i+1], 14)
            DOD_mat.append(val)



        #BOD

        if df[paramA[4]][i] < df[paramA[4]][i+1]:
                val = np.random.uniform(df[paramA[4]][i+1], df[paramA[4]][i], 14)
                BODA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[4]][i], df[paramA[4]][i+1], 14)
            BODA_mat.append(val)

        if df[paramB[4]][i] < df[paramB[4]][i+1]:
                val = np.random.uniform(df[paramB[4]][i+1], df[paramB[4]][i], 14)
                BODB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[4]][i], df[paramB[4]][i+1], 14)
            BODB_mat.append(val)

        if df[paramC[4]][i] < df[paramC[4]][i+1]:
                val = np.random.uniform(df[paramC[4]][i+1], df[paramC[4]][i], 14)
                BODC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[4]][i], df[paramC[4]][i+1], 14)
            BODC_mat.append(val)

        if df[paramD[4]][i] < df[paramD[4]][i+1]:
                val = np.random.uniform(df[paramD[4]][i+1], df[paramD[4]][i], 14)
                BODD_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[4]][i], df[paramD[4]][i+1], 14)
            BODD_mat.append(val)


        #TN

        if df[paramA[5]][i] < df[paramA[5]][i+1]:
                val = np.random.uniform(df[paramA[5]][i+1], df[paramA[5]][i], 14)
                TNA_mat.append(val)
        else:
            val = np.random.uniform(df[paramA[5]][i], df[paramA[5]][i+1], 14)
            TNA_mat.append(val)

        if df[paramB[5]][i] < df[paramB[5]][i+1]:
                val = np.random.uniform(df[paramB[5]][i+1], df[paramB[5]][i], 14)
                TNB_mat.append(val)
        else:
            val = np.random.uniform(df[paramB[5]][i], df[paramB[5]][i+1], 14)
            TNB_mat.append(val)

        if df[paramC[5]][i] < df[paramC[5]][i+1]:
                val = np.random.uniform(df[paramC[5]][i+1], df[paramC[5]][i], 14)
                TNC_mat.append(val)
        else:
            val = np.random.uniform(df[paramC[5]][i], df[paramC[5]][i+1], 14)
            TNC_mat.append(val)

        if df[paramD[5]][i] < df[paramD[5]][i+1]:
                val = np.random.uniform(df[paramD[5]][i+1], df[paramD[5]][i], 14)
                TND_mat.append(val)
        else:
            val = np.random.uniform(df[paramD[5]][i], df[paramD[5]][i+1], 14)
            TND_mat.append(val)

        i+=1
    
    return tempA_mat,tempB_mat,tempC_mat,tempD_mat,pHA_mat,pHB_mat,pHC_mat,pHD_mat,ECA_mat,ECB_mat,ECC_mat,ECD_mat,DOA_mat,DOB_mat,DOC_mat,DOD_mat,BODA_mat,BODB_mat,BODC_mat,BODD_mat,TNA_mat, TNB_mat, TNC_mat, TND_mat
        
```
